```python
import json
from collections import defaultdict

def organize_labels(file_path):
    with open(file_path, 'r', encoding='utf-8') as f:
        data = json.load(f)

    summary = defaultdict(lambda: {"count": 0, "total_duration": 0})
    organized_results = []

    for entry in data:
        video_info = {
            "video_path": entry.get("video"),
            "annotation_id": entry.get("annotation_id"),
            "labels": []
        }
        
        # 遍历视频中的所有标注
        labels_list = entry.get("videoLabels", [])
        for label_item in labels_list:
            # 获取标签名称（通常列表里只有一个元素）
            label_names = label_item.get("timelinelabels", [])
            ranges = label_item.get("ranges", [])
            
            for r in ranges:
                start = r.get("start")
                end = r.get("end")
                duration = end - start
                
                for name in label_names:
                    # 添加到该视频的整理列表中
                    video_info["labels"].append({
                        "label": name,
                        "start": start,
                        "end": end,
                        "duration": duration
                    })
                    
                    # 更新全局统计信息
                    summary[name]["count"] += 1
                    summary[name]["total_duration"] += duration
        
        # 按开始时间排序
        video_info["labels"].sort(key=lambda x: x['start'])
        organized_results.append(video_info)

    return organized_results, summary

# --- 执行整理 ---
results, stats = organize_labels(mini_path)

# --- 打印整理后的结果 ---
print("=== 标签统计摘要 ===")
print(f"{'标签':<15} | {'出现次数':<10} | {'总持续时间'}")
print("-" * 45)
for label, info in stats.items():
    print(f"{label:<15} | {info['count']:<10} | {info['total_duration']}")

print("\n=== 视频详细标注明细 (前5条) ===")
for vid in results[:1]:  # 示例打印第一个视频
    print(f"\n视频文件: {vid['video_path']}")
    for l in vid['labels'][:10]:  # 示例打印前10个标签
        print(f"  [{l['start']:>5} - {l['end']:>5}] 动作: {l['label']:<10} (时长: {l['duration']})")
    print("  ...")
```

```python
# 有些annotation只打了4个标签，有些打了8个标签。需要统计一下

#!/usr/bin/env python3
# -*- coding: utf-8 -*-

import json
from pathlib import Path
from collections import Counter, defaultdict

def iter_labels_in_annotation(ann: dict):
    """Yield timelinelabel strings inside one annotation item."""
    for vl in ann.get("videoLabels", []) or []:
        for tl in (vl.get("timelinelabels", []) or []):
            if isinstance(tl, str) and tl.strip():
                yield tl.strip()

def scan(label_dir: Path):
    results = []
    for json_path in sorted(label_dir.rglob("*.json")):
        # robust read (utf-8 / utf-8-sig)
        try:
            text = json_path.read_text(encoding="utf-8")
            data = json.loads(text)
        except UnicodeDecodeError:
            text = json_path.read_text(encoding="utf-8-sig")
            data = json.loads(text)
        except Exception as e:
            print(f"[WARN] failed to read {json_path}: {e}")
            continue

        # file-level: annotator -> Counter(label -> count)
        per_annotator_counts: dict[str, Counter] = defaultdict(Counter)

        for ann in data.get("annotations", []) or []:
            aid = str(ann.get("annotator", "UNKNOWN"))
            for lb in iter_labels_in_annotation(ann):
                per_annotator_counts[aid][lb] += 1

        # make JSON-serializable summary
        per_annotator_summary = {}
        for aid, c in per_annotator_counts.items():
            per_annotator_summary[aid] = {
                "n_classes": len(c),
                "classes": sorted(c.keys()),
                "counts": dict(c.most_common()),   # label->count
                "n_total": int(sum(c.values())),
            }

        # union (optional)
        union_classes = sorted({lb for c in per_annotator_counts.values() for lb in c.keys()})

        results.append({
            "file": str(json_path.relative_to(label_dir.parent)),
            "annotators": sorted(per_annotator_summary.keys(), key=lambda x: (x=="UNKNOWN", x)),
            "per_annotator": per_annotator_summary,
            "union_classes": union_classes,
            "n_union_classes": len(union_classes),
        })

    return results


label_dir = Path("/workspace/data/multi_view_driver_action/label")
if not label_dir.exists():
    raise SystemExit(f"label folder not found: {label_dir}")

results = scan(label_dir)

# pretty print to console
for r in results:
    print(f"\n=== {r['file']} ===")
    for aid in r["annotators"]:
        s = r["per_annotator"][aid]
        print(f"- annotator {aid}: classes={s['n_classes']} total={s['n_total']}")
        print(f"  classes: {s['classes']}")
        print(f"  counts:  {s['counts']}")
    print(f"  union({r['n_union_classes']}): {r['union_classes']}")

out_path = "per_file_annotator_label_stats.json"
with open(out_path, 'w', encoding='utf-8') as out_f:
    json.dump(results, out_f, indent=4, ensure_ascii=False)
print(f"\n[INFO] saved summary to {out_path}")
```

```python
# 统计每个视频的标签分布情况，检查标签不平衡
import json
import math
from pathlib import Path
from collections import Counter
import csv

def extract_labels_and_duration(annotation_item: dict):
    """从单条 annotation 中提取标签计数与持续时长。"""
    label_counter = Counter()
    duration_counter = Counter()

    for vl in annotation_item.get("videoLabels", []) or []:
        labels = [lb.strip() for lb in (vl.get("timelinelabels", []) or []) if isinstance(lb, str) and lb.strip()]
        ranges = vl.get("ranges", []) or []

        for r in ranges:
            start = r.get("start")
            end = r.get("end")
            if start is None or end is None:
                continue
            duration = max(float(end) - float(start), 0.0)
            for lb in labels:
                label_counter[lb] += 1
                duration_counter[lb] += duration

    return label_counter, duration_counter

def calc_entropy(counter: Counter) -> float:
    total = sum(counter.values())
    if total <= 0:
        return 0.0
    entropy = 0.0
    for c in counter.values():
        p = c / total
        if p > 0:
            entropy -= p * math.log(p, 2)
    return entropy

def imbalance_ratio(counter: Counter) -> float:
    """最大类/最小类（最小类为0时不参与，避免除零）。"""
    vals = [v for v in counter.values() if v > 0]
    if not vals:
        return float("inf")
    return max(vals) / min(vals)

# 你可以改这里：
label_dir = Path("/workspace/data/multi_view_driver_action/label")
out_json = Path("/workspace/MultiView_DriverAction_PyTorch/analysis/per_video_label_distribution.json")
out_csv = Path("/workspace/MultiView_DriverAction_PyTorch/analysis/per_video_label_distribution.csv")

if not label_dir.exists():
    raise SystemExit(f"label folder not found: {label_dir}")

records = []
all_label_set = set()

for json_path in sorted(label_dir.rglob("*.json")):
    try:
        data = json.loads(json_path.read_text(encoding="utf-8"))
    except UnicodeDecodeError:
        data = json.loads(json_path.read_text(encoding="utf-8-sig"))
    except Exception as e:
        print(f"[WARN] failed to read {json_path}: {e}")
        continue

    video_file = data.get("video_file", json_path.stem)
    annotations = data.get("annotations", []) or []

    cnt_total = Counter()
    dur_total = Counter()

    for ann in annotations:
        c, d = extract_labels_and_duration(ann)
        cnt_total.update(c)
        dur_total.update(d)

    all_label_set.update(cnt_total.keys())

    n_classes = len(cnt_total)
    n_segments = int(sum(cnt_total.values()))
    total_duration = float(sum(dur_total.values()))
    ratio = imbalance_ratio(cnt_total) if n_classes > 1 else 1.0
    ent = calc_entropy(cnt_total)

    records.append({
        "video": video_file,
        "num_annotators": len(annotations),
        "num_classes": n_classes,
        "num_segments": n_segments,
        "total_duration": total_duration,
        "imbalance_ratio_max_min": ratio,
        "label_entropy": ent,
        "label_count": dict(cnt_total),
        "label_duration": {k: round(v, 4) for k, v in dur_total.items()},
    })

# 控制台摘要（按不平衡程度从高到低）
records_sorted = sorted(records, key=lambda x: x["imbalance_ratio_max_min"], reverse=True)
print(f"videos: {len(records_sorted)}")
print("\nTop-10 imbalance videos (max/min ratio):")
for r in records_sorted[:10]:
    print(
        f"- {r['video']}: ratio={r['imbalance_ratio_max_min']:.3f}, "
        f"classes={r['num_classes']}, segments={r['num_segments']}, entropy={r['label_entropy']:.3f}"
    )

# 写 JSON
out_json.parent.mkdir(parents=True, exist_ok=True)
with out_json.open("w", encoding="utf-8") as f:
    json.dump(records_sorted, f, ensure_ascii=False, indent=2)

# 写 CSV（展开 label_count）
label_columns = sorted(all_label_set)
fieldnames = [
    "video",
    "num_annotators",
    "num_classes",
    "num_segments",
    "total_duration",
    "imbalance_ratio_max_min",
    "label_entropy",
] + [f"count::{lb}" for lb in label_columns]

with out_csv.open("w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for r in records_sorted:
        row = {
            "video": r["video"],
            "num_annotators": r["num_annotators"],
            "num_classes": r["num_classes"],
            "num_segments": r["num_segments"],
            "total_duration": round(r["total_duration"], 4),
            "imbalance_ratio_max_min": round(r["imbalance_ratio_max_min"], 6),
            "label_entropy": round(r["label_entropy"], 6),
        }
        for lb in label_columns:
            row[f"count::{lb}"] = r["label_count"].get(lb, 0)
        writer.writerow(row)

print(f"\n[INFO] saved JSON: {out_json}")
print(f"[INFO] saved CSV : {out_csv}")
```

```python
# 8分类 -> 4分类后，重算每个视频的标签分布与不平衡指标
import json
import math
import csv
from pathlib import Path
from collections import Counter

# 4分类映射
MAP_8_TO_4 = {
    "left": "left",
    "left_up": "left",
    "left_down": "left",
    "right": "right",
    "right_up": "right",
    "right_down": "right",
    "up": "up",
    "down": "down",
}

def to4(label: str) -> str:
    return MAP_8_TO_4.get(str(label), str(label))

def calc_entropy(counter: Counter) -> float:
    total = sum(counter.values())
    if total <= 0:
        return 0.0
    entropy = 0.0
    for c in counter.values():
        p = c / total
        if p > 0:
            entropy -= p * math.log(p, 2)
    return entropy

def imbalance_ratio(counter: Counter) -> float:
    vals = [v for v in counter.values() if v > 0]
    if not vals:
        return float("inf")
    return max(vals) / min(vals)

src_json = Path("/workspace/MultiView_DriverAction_PyTorch/analysis/per_video_label_distribution.json")
out_json_4 = Path("/workspace/MultiView_DriverAction_PyTorch/analysis/per_video_label_distribution_4class.json")
out_csv_4 = Path("/workspace/MultiView_DriverAction_PyTorch/analysis/per_video_label_distribution_4class.csv")

if not src_json.exists():
    raise FileNotFoundError(f"未找到源统计文件: {src_json}")

with src_json.open("r", encoding="utf-8") as f:
    records = json.load(f)

records_4 = []
for r in records:
    cnt_raw = Counter(r.get("label_count", {}))
    dur_raw = Counter(r.get("label_duration", {}))

    cnt_4 = Counter()
    dur_4 = Counter()
    for lb, c in cnt_raw.items():
        cnt_4[to4(lb)] += int(c)
    for lb, d in dur_raw.items():
        dur_4[to4(lb)] += float(d)

    n_classes_4 = len([k for k, v in cnt_4.items() if v > 0])
    n_segments_4 = int(sum(cnt_4.values()))
    total_duration_4 = float(sum(dur_4.values()))
    ratio_4 = imbalance_ratio(cnt_4) if n_classes_4 > 1 else 1.0
    ent_4 = calc_entropy(cnt_4)

    records_4.append({
        "video": r.get("video"),
        "num_annotators": r.get("num_annotators", None),
        "num_classes": n_classes_4,
        "num_segments": n_segments_4,
        "total_duration": total_duration_4,
        "imbalance_ratio_max_min": ratio_4,
        "label_entropy": ent_4,
        "label_count": dict(cnt_4),
        "label_duration": {k: round(v, 4) for k, v in dur_4.items()},
    })

records_4_sorted = sorted(records_4, key=lambda x: x["imbalance_ratio_max_min"], reverse=True)

out_json_4.parent.mkdir(parents=True, exist_ok=True)
with out_json_4.open("w", encoding="utf-8") as f:
    json.dump(records_4_sorted, f, ensure_ascii=False, indent=2)

label_columns_4 = ["left", "right", "up", "down"]
fieldnames = [
    "video",
    "num_annotators",
    "num_classes",
    "num_segments",
    "total_duration",
    "imbalance_ratio_max_min",
    "label_entropy",
] + [f"count::{lb}" for lb in label_columns_4]

with out_csv_4.open("w", newline="", encoding="utf-8") as f:
    writer = csv.DictWriter(f, fieldnames=fieldnames)
    writer.writeheader()
    for r in records_4_sorted:
        row = {
            "video": r["video"],
            "num_annotators": r["num_annotators"],
            "num_classes": r["num_classes"],
            "num_segments": r["num_segments"],
            "total_duration": round(r["total_duration"], 4),
            "imbalance_ratio_max_min": round(r["imbalance_ratio_max_min"], 6),
            "label_entropy": round(r["label_entropy"], 6),
        }
        for lb in label_columns_4:
            row[f"count::{lb}"] = r["label_count"].get(lb, 0)
        writer.writerow(row)

print(f"videos: {len(records_4_sorted)}")
print("\nTop-10 imbalance videos after 4-class merge:")
for rr in records_4_sorted[:10]:
    print(
        f"- {rr['video']}: ratio={rr['imbalance_ratio_max_min']:.3f}, "
        f"classes={rr['num_classes']}, segments={rr['num_segments']}, entropy={rr['label_entropy']:.3f}"
    )

print(f"\n[INFO] saved JSON: {out_json_4}")
print(f"[INFO] saved CSV : {out_csv_4}")
```

```python
# 4分类占比堆叠条形图（每个视频）
import json
from pathlib import Path
import pandas as pd
import matplotlib.pyplot as plt

json_path_4 = Path("/workspace/MultiView_DriverAction_PyTorch/analysis/per_video_label_distribution_4class.json")
fig_dir_4 = Path("/workspace/MultiView_DriverAction_PyTorch/analysis/figures_4class")
fig_dir_4.mkdir(parents=True, exist_ok=True)

if not json_path_4.exists():
    raise FileNotFoundError(f"4分类统计结果不存在，请先运行4分类统计单元: {json_path_4}")

with json_path_4.open("r", encoding="utf-8") as f:
    rows_4 = json.load(f)

if len(rows_4) == 0:
    raise ValueError("4分类统计结果为空。")

# 展开 label_count -> left/right/up/down 比例
classes = ["left", "right", "up", "down"]
plot_rows = []
for r in rows_4:
    counts = r.get("label_count", {})
    total = sum(float(counts.get(c, 0)) for c in classes)
    if total <= 0:
        continue
    item = {"video": r.get("video", "unknown")}
    for c in classes:
        item[c] = float(counts.get(c, 0)) / total
    item["imbalance_ratio_max_min"] = float(r.get("imbalance_ratio_max_min", 0.0))
    plot_rows.append(item)

df_prop = pd.DataFrame(plot_rows)
if df_prop.empty:
    raise ValueError("没有可绘制的视频比例数据。")

# 为了可读性，按不平衡程度排序，只画 Top-N
top_n = min(20, len(df_prop))
df_top = df_prop.sort_values("imbalance_ratio_max_min", ascending=False).head(top_n).copy()
df_top = df_top.set_index("video")

plt.figure(figsize=(16, 7))
bottom = None
colors = {
    "left": "#1f77b4",
    "right": "#ff7f0e",
    "up": "#2ca02c",
    "down": "#d62728",
}
for c in classes:
    vals = df_top[c].values
    if bottom is None:
        plt.bar(range(len(df_top)), vals, label=c, color=colors[c])
        bottom = vals
    else:
        plt.bar(range(len(df_top)), vals, bottom=bottom, label=c, color=colors[c])
        bottom = bottom + vals

plt.xticks(range(len(df_top)), df_top.index.tolist(), rotation=75, ha="right")
plt.ylabel("Class Proportion")
plt.ylim(0, 1.0)
plt.title(f"[4-class] Per-video Label Proportion (Stacked, Top-{top_n} by Imbalance)")
plt.legend(title="Class", loc="upper right")
plt.tight_layout()

p4_4 = fig_dir_4 / "per_video_class_proportion_stacked_4class.png"
plt.savefig(p4_4, dpi=220)
plt.show()

print(f"[INFO] 已保存: {p4_4}")
```
